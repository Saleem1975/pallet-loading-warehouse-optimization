# Pallet Loading Optimization Demo

This notebook demonstrates how to work with the summary data prepared for the pallet loading optimization project.

It covers:

- Dynamic compression strength factors
- Pallet volume utilization comparisons
- Relative humidity and load height
- Storage time and load height


```
import pandas as pd
import matplotlib.pyplot as plt

pvu = pd.read_csv("../data/selected_pvu_examples.csv")
humidity = pd.read_csv("../data/selected_humidity_impact_examples.csv")
storage = pd.read_csv("../data/selected_storage_time_impact_examples.csv")
storage_factors = pd.read_csv("../data/storage_time_factors.csv")
humidity_factors = pd.read_csv("../data/relative_humidity_factors.csv")

pvu.head()
```

## Pallet volume utilization improvement

```
top_gain = pvu.sort_values("PVU_gain_1_to_best", ascending=False).head(10)

plt.figure(figsize=(10, 5))
plt.bar(top_gain["dataset"].astype(str), top_gain["PVU_gain_1_to_best"])
plt.title("Largest PVU improvement from mixed layer types")
plt.xlabel("Dataset")
plt.ylabel("PVU improvement points")
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

## Relative humidity factor

```
humidity_factors
```

```
rh_mid = [22.5, 50, 60, 70, 80, 92.5]

plt.figure(figsize=(9, 5))
plt.plot(rh_mid, humidity_factors["FH_relative_humidity_factor"], marker="o")
plt.title("Relative humidity factor")
plt.xlabel("Relative humidity midpoint (%)")
plt.ylabel("FH factor")
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

## Storage time factor

```
storage_factors
```

```
storage_mid = [0, 2, 7, 20.5, 60.5, 105.5, 210.5]

plt.figure(figsize=(9, 5))
plt.plot(storage_mid, storage_factors["FT_storage_time_factor"], marker="o")
plt.title("Storage time factor")
plt.xlabel("Representative storage time midpoint (days)")
plt.ylabel("FT factor")
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

## Average load height under humidity scenarios

```
avg_humidity = humidity[
    ["load_height_RH_40", "load_height_RH_71", "load_height_RH_79_84", "load_height_RH_90"]
].mean()

plt.figure(figsize=(9, 5))
plt.plot(["40%", "71%", "79–84%", "90%"], avg_humidity.values, marker="o")
plt.title("Average load height under humidity scenarios")
plt.xlabel("Relative humidity")
plt.ylabel("Average load height")
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

## Average load height under storage-time scenarios

```
avg_storage = storage[
    ["load_height_10_days", "load_height_3_months", "load_height_4_months"]
].mean()

plt.figure(figsize=(9, 5))
plt.plot(["10 days", "3 months", "4 months"], avg_storage.values, marker="o")
plt.title("Average load height under storage-time scenarios")
plt.xlabel("Storage time")
plt.ylabel("Average load height")
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```
